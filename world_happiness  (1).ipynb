## Data cleaning and analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
%matplotlib inline
```

First let's import csv files, for years that are available. Than I will enlist the column names for all years and tray to get them as uniform as it's possible across the datasets. This will make it easier to exam data and to marge them in one dataset.

```python
df_15 = pd.read_csv('2015.csv')
df_16 = pd.read_csv('2016.csv')
df_17 = pd.read_csv('2017.csv')
df_18 = pd.read_csv('2018.csv')
df_19 = pd.read_csv('2019.csv')
```

```python
df = [df_15, df_16, df_17, df_18, df_19]

for frames in df:
    print(frames.columns)
    print('/n')
```

```python
#drop columns across datasets that are different
col_name = ['Region', 'Standard Error', 'Dystopia Residual', 'Dystopia.Residual', 'Lower Confidence Interval', 'Upper Confidence Interval','Whisker.high',
       'Whisker.low', 'Social support']
for frame in df:
    for col in col_name:
        if col in frame:
            frame.drop(col, axis = 1, inplace = True)
```

```python
for frame in df:
    print(frame.columns)
    print('/n')
```

```python
#change names of columns in order to be uniform accros different datasets
for frame in df:
    frame.rename(columns = {'Country or region': 'Country', 'Happiness.Rank':'Happiness Rank','Overall rank':'Happiness Rank', 'Happiness.Score':'Happiness Score', 'Score':'Happiness Score', 'Economy..GDP.per.Capita.':'Economy (GDP per Capita)', 'GDP per capita':'Economy (GDP per Capita)', 'Health..Life.Expectancy.':'Health (Life Expectancy)', 'Healthy life expectancy':'Health (Life Expectancy)', 'Freedom to make life choices':'Freedom', 'Trust..Government.Corruption.':'Trust (Government Corruption)', 'Perceptions of corruption':'Trust (Government Corruption)'}, inplace = True)
```

```python
for frame in df:
    print(frame.columns)
    print('/n')
```

For reshaping datasets I will write a funcion reshape_frame.

```python
for frame in df[2:]:
    frame_1 = frame['Generosity']
    frame.drop(columns = ['Generosity'], inplace = True)
    frame.insert(len(frame.columns), 'Generosity', frame_1)
```

```python
def reshape_frame(df, column_name, pos):
    for frame in df:
        frame_1 = frame[column_name]
        frame.drop(columns = [column_name], inplace = True)
        frame.insert(pos, column_name, frame_1)   
```

```python
reshape_frame(df[3:], 'Happiness Rank', 1)
```

```python
for frame in df:
    print(frame.columns)
    print('/n')
```

```python
df_15['Year'] = 2015
df_16['Year'] = 2016
df_17['Year'] = 2017
df_18['Year'] = 2018
df_19['Year'] = 2019
```
